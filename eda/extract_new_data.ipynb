```python
dfs = [df_cloud_cover, df_sunshine, df_global_radiation, df_max_temp, df_mean_temp, df_min_temp, df_precipitation, df_pressure, df_snow_depth]
```

```python
import functools as ft
df_final = ft.reduce(lambda left, right: pd.merge(left, right, on='date'), dfs)
```

```python
df_final.head()
```

```python
df_final=df_final.drop(columns=['cc', 'ss', 'qq', 'tx', 'tg', 'tn', 'rr','pp', 'sd'])
```

```python
df_final.head()
```

```python
df_final.tail()
```

```python
df_final.dtypes
```

```python
df_final['cloud_cover'] = df_final['cloud_cover'].astype(float)
df_final['global_radiation'] = df_final['global_radiation'].astype(float)
df_final['pressure'] = df_final['pressure'].astype(float)
df_final['snow_depth'] = df_final['snow_depth'].astype(float)
```

```python
df_final.head()
```

```python
df_final.describe()
```

```python
# We will fix those columns where the NaN has been loaded as upper 9s
df_final['cloud_cover'].replace(-9999.0, np.nan, inplace=True)
df_final['global_radiation'].replace(-9999.0, np.nan, inplace=True)
df_final['mean_temp'].replace(-999.9, np.nan, inplace=True)
df_final['pressure'].replace(-99990.0, np.nan, inplace=True)
df_final['snow_depth'].replace(-9999.0, np.nan, inplace=True)
```

```python
df_final.describe()
```

```python
mask = (df_final['date'] >= 19790101) & (df_final['date'] <= 20201231)
print(df_final.loc[mask])
df_subset = df_final.loc[mask]
```
